```python
from dotenv import load_dotenv
import os

# Load variables from .env
load_dotenv()

# Get the API key
cohere_api_key = os.getenv("cohere_api_key")
```

## Conclusion

Among the three reranking techniques, **Cohere API Reranker** produced the best results for this query. It ranked the most relevant documents with the highest confidence by understanding the semantic meaning and context of the query.

The **Cross-Encoder** also performed well and is an excellent choice for local, high-quality reranking without relying on an external API.

**BM25** is the fastest approach and works well for keyword matching, but it cannot capture semantic relationships as effectively as Cross-Encoder or Cohere.

**Overall Choice:** **Cohere API Reranker** because it provides the highest-quality reranking, making it the preferred option for production-grade RAG applications where retrieval accuracy is critical.
